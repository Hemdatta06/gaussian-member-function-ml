```python
import numpy as np
import matplotlib.pyplot as plt

heights = np.linspace(145, 180, 15)

frequencies = np.array([1, 2, 2, 4, 5, 8, 11, 13, 9, 6, 5, 3, 2, 1, 1])

mean = np.average(heights, weights=frequencies)
sigma = np.sqrt(np.average((heights - mean)**2, weights=frequencies))

def gaussian_mf(x, mean, sigma):
    return np.exp(-0.5 * ((x - mean) / sigma)**2)

fig, ax1 = plt.subplots(figsize=(10, 6))

bar_width = heights[1] - heights[0]
ax1.bar(heights, frequencies, width=bar_width, alpha=0.7,
        color='#e9c46a', edgecolor='#264653', linewidth=1.2)

ax1.set_xlabel("Height (cm)", fontsize=11)
ax1.set_ylabel("Frequency (Number of Students)", color='#264653', fontsize=11)

ax2 = ax1.twinx()
x_smooth = np.linspace(140, 185, 500)
y_membership = gaussian_mf(x_smooth, mean, sigma)

ax2.plot(x_smooth, y_membership, color='#e76f51', linewidth=2.5, label='Fuzzy MF')
ax2.set_ylabel("Degree of Membership", color='#e76f51', fontsize=11)
ax2.set_ylim(0, 1.1)
plt.title("Gaussian Membership Function Analysis", pad=20, fontweight='bold')
ax1.axvline(mean, color='#2a9d8f', linestyle='--', linewidth=2, label=f'Mean: {mean:.1f}')

ax1.grid(True, axis='y', linestyle='-', alpha=0.2)
ax1.set_axisbelow(True)

plt.show()

print(f"Total Students: {np.sum(frequencies)}")
print(f"Mean: {mean:.2f} | Sigma: {sigma:.2f}")
```
